```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
TE = 0.01
nsamp = 100

t = np.linspace(0.5*TE, 1.5*TE, nsamp)[:,None]
t = np.repeat(t, nsamp, axis=1)
t = np.flip(t, axis=0)

def plot_arr(arr):
    plt.imshow(arr, cmap='gray', extent=(0, nsamp, 0, nsamp))
    plt.show()

def plot_arr_horiz(arrays):
    """
    Plots a list of arrays horizontally using matplotlib's imshow in separate subplots.

    Parameters:
    arrays (list of np.ndarray): List of arrays to be plotted.
    """
    # Check if the list is empty
    if not arrays:
        raise ValueError("The list of arrays is empty")

    # Determine the number of subplots
    num_subplots = len(arrays)

    # Create a figure and a set of subplots
    fig, axes = plt.subplots(1, num_subplots, figsize=(num_subplots * 5, 5))

    # If there's only one subplot, axes is not a list, so we need to make it a list
    if num_subplots == 1:
        axes = [axes]

    # Plot each array in a separate subplot
    for ax, array in zip(axes, arrays):
        ax.imshow(array, aspect='auto')
        ax.axis('off')  # Turn off axis

    # Adjust layout
    plt.tight_layout()
    plt.show()
```

```python
rate_map = np.random.rand(nsamp) + 1j*np.random.rand(nsamp)
rate_map = np.repeat(rate_map[None,:], nsamp, axis=0)

img = np.random.rand(nsamp) + 1j*np.random.rand(nsamp)
img = np.repeat(img[None,:], nsamp, axis=0)

plot_arr_horiz([t, img.real, img.imag])

offres = np.exp(-rate_map*t)

plot_arr_horiz([rate_map.real, rate_map.imag])
plot_arr_horiz([offres.real, offres.imag])

```
